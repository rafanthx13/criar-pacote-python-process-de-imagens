# Testando meu próprio pacote do Pypi

Comandos para subir o pacote:

+ 1 - Saiba o user/senha da sua conta
+ 2 - Criar a pasta `dist/` que tem os arquivos a serem enviados
  - ```
  python setup.py sdist bdist_wheel
  ```
+ 3 - Mandar para o pypi
  - ```
  python -m twine upload --repository-url https://upload.pypi.org/legacy/ dist/*
  ```
